## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os
import sys
import subprocess
import pandas as pd

# -----------------------------
# Setup
# -----------------------------
IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-q", "-r", "requirements.txt"],
        check=True
    )
else:
    while not os.path.isdir("data/raw") and os.getcwd() != "/":
        os.chdir("..")

print("Working directory:", os.getcwd())

assert os.path.exists(
    "data/raw/content_refresh_anonymized.csv"
), "Starter CSV not found."

# -----------------------------
# Load the starter dataset
# -----------------------------
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# -----------------------------
# Quick look at the data
# -----------------------------
total_pages = len(df)
trend_counts = df["trend_direction"].value_counts()
median_position = df["avg_position"].median()
min_position = df["avg_position"].min()
max_position = df["avg_position"].max()

print("\n=== Quick Look at the Data ===\n")

print(
    f"1. The starter dataset contains {total_pages:,} content pages, "
    "which is large enough that manually reviewing every page is impractical. "
    "This makes a ranked review system valuable.\n"
)

print(
    f"2. The trend distribution shows that {trend_counts['down']:,} pages are "
    f"labeled as 'down', while only {trend_counts['up']:,} are labeled as 'up'. "
    "This suggests there are many pages that may require investigation, making "
    "prioritization an important task.\n"
)

print(
    f"3. The median average search position is {median_position:.1f}, "
    f"with values ranging from {min_position:.0f} to {max_position:.0f}. "
    "This wide variation in visibility indicates that search performance "
    "differs considerably across pages and can be used as an important "
    "signal when deciding which pages should be reviewed first."
)
```
